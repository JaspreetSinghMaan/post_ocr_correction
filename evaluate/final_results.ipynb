```python
import pandas as pd
import glob
```

```python
pd.set_option("precision", 2)
```

```python
files = glob.glob("csv/*.csv")
files
```

```python
data = [pd.read_csv(f).assign(language = f) for f in files]
data = pd.concat(data)
data.head(20)
```

```python
data\
.sort_values(["window", "window_size", "decoding", "weighting"])\
.pivot(index = "language", 
       values = "improvement", 
       columns = ["window", "window_size", "decoding", "weighting"])
```

```python
data\
.sort_values(["window", "window_size", "decoding", "weighting"])\
.pivot(index = "language", 
       values = "inference_seconds", 
       columns = ["window", "window_size", "decoding", "weighting"]) / 60
```
